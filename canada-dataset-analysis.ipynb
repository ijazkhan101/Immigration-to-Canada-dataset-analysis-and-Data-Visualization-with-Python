```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import matplotlib as mpl
import seaborn as sns
```

```python
df_can = pd.read_csv('canadian_immegration_data.csv')
df_can.head()
```

```python
df_can.isnull().sum()
```

```python
df_can.describe()
```

```python
# filtering on the list of countries ('Country')
df_can.Country
```

```python
# converting the column names into strings
df_can.columns = list(map(str, df_can.columns))
years = list(map(str, range(1980, 2014)))
years
```

```python
df_can[(df_can['Continent']=='Asia') & (df_can['Region']=='Eastern Asia')]
```

```python
print('data dimensions:', df_can.shape)
print(df_can.columns)
df_can.head(2)
```

```python
df_continent = df_can.groupby('Continent', axis = 0).sum()
df_continent['Total'].plot(kind = 'pie', figsize=(10,5), shadow=True, autopct='%1f%%') # autopct create %
plt.title('Immigrants by Continent to Canada')
plt.axis('equal')
plt.show()

```

### Top 5 Countries of Immigrants

```python


df_can_top5 = df_can.nlargest(n=5, columns='Total')
df_can_top5 = df_can_top5.groupby("Country", axis=0).sum()
df_can_top5.head()
df_can_top5['Total'].sort_values(ascending=False).plot(kind='bar', figsize=(10, 5), rot=90, color='lightcoral')
plt.title('Top 5 Countries of Immigrants')
plt.ylabel('Number of immigrants')
plt.xticks(rotation=90)

```

```python
inplace = True 
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].transpose() 
print(df_top5)

df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14,10),color="rgbcm") 

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

### Bottom 5 Countries of Immigrants

```python
df_can_btm5 = df_can.nsmallest(n=5, columns='Total')
df_can_btm5 = df_can_btm5.groupby("Country", axis=0).sum()
df_can_btm5.head()
df_can_btm5['Total'].sort_values(ascending=False).plot(kind='bar', figsize=(10, 5), rot=90, color='Coral')
plt.title('Bottom 5 Countries of Immigrants')
plt.ylabel('Number of immigrants')
plt.xticks(rotation=0)

```

### Histogram of Immigration from 195 countries in 2013

```python
df_can['2013'].head()
# plotting histogram for 2013 new immigrants to Canada

count, bin_edges = np.histogram(df_can['2013'])

df_can['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges,color="darkgreen")

plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries') 
plt.xlabel('Number of Immigrants')
plt.show()
```
